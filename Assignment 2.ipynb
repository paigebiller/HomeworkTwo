Create a program that prompts for your weight, gender, number of drinks, the amount of alcohol by volume of the drinks consumed, and the amount of time since your last drink. Calculate your blood alcohol content (BAC) using this formula $BAC = (A \times 5.14 / W \times r) - 0.015 \times H$

where

- A is total alcohol consumed, in ounces (oz).
- W is body weight in pounds.
- r is the alcohol distribution ratio:
  - 0.73 for men
  - 0.66 for women
- H is number of hours since the last drink.

Display whether or not it’s legal to drive by comparing the blood alcohol content to 0.08.

## Example Output
<pre>

</pre>

## Constraint

```python
#try statement prevents user from entering non-numeric values
#gender is coded as 2 for Male and 1 for female to also benefit
#from this exception check
#BAC calculated from inputs given
#verdict variable declared as '' if BAC is less than 0.08, otherwise 'not'
#so that a single print statement can be used

def problem4():
    while True:
        try:
            weight = float(input("What is your weight in pounds? "))
            gender=int(input("If Male, enter 2. If Female, enter 1. "))
            alcohol=float(input("How many ounces of alcohol have you drank? "))
            hourssince=float(input("How many hours since your last drink? "))
        except Exception as e:
            print("Please enter only numeric values.")
            print(e)
            continue
        break

    verdict=' not'

    if gender==2:
        alcoholdistribratio=0.73
    else:
        alcoholdistribratio=0.66

    BAC = (alcohol*5.14/weight * alcoholdistribratio) - (0.015 * hourssince)

    if BAC < 0.08:
        verdict=''

    print('Your BAC is {BAC:.}.\n'
          'It is{verdict} legal for you to drive.'.format(verdict=verdict, BAC=BAC))

problem4()
```

## Example Output

## Constraints

- Implement the program using a function called `isAnagram`, which takes in two words as its arguments and returns `true` or `false`. Invoke this function from your main program.
- Check that both words are the same length.

```python
#this program calls a function within a function to determine whether two string are anagrams
#meaning they share the same letters and the same string length, which is determined with len()
#the isanagram function returns true or false, which determines what ourput prints in function problem9()

def problem9():
    string1=input('Enter two strings and I\'ll tell you if they are anagrams: \n Enter the first string: ')
    string2=input('Enter the second string: ')

    def isAnagram(string1, string2):
        string1 = sorted(string1)
        string2 = sorted(string2)
        if string1 == string2 and len(string1)==len(string2):
            return True
        else:
            return False

    if isAnagram(string1,string2)==True:
        print('"{string1}" and "{string2}" are anagrams.'.format(string1=string1, string2=string2))
    else:
        print('"{string1}" and "{string2}" are not anagrams.'.format(string1=string1, string2=string2))

problem9()
```

# Problem 10 (4 points)

Create a program that determines the complexity of a given password based on these rules:
- A very weak password contains only numbers and is fewer than eight characters.
- A weak password contains only letters and is fewer than eight characters.
- A strong password contains letters and at least one number and is at least eight characters.
- A very strong password contains letters, numbers, and special characters and is at least eight characters.

## Example Output

<pre>
The password '12345' is a very weak password.
The password 'abcdef' is a weak password.
The password 'abc123xyz' is a strong password.

## Constraints

- Create a `passwordValidator` function that takes in the password as its argument and returns a value you can evaluate to determine the password strength. Do not have the function return a string — you may need to support multiple languages in the future.
- Use a single output statement.

```python
#the function determines password strength by using the string attributes .isdigit and .isalpha
#to determine whether all or any of the chacaters in the password string are alphabetic or numeric
#and returns digits instead of strings, which are then assigning string sentence fragments so that
#only one output statement needs to be used
#any string not in any of the four defined categories prints a request to enter a different password

def problem10():
    password = input('Enter a password to test: ')
    def passwordValidator(password):
        if len(password)<8 and \
        password.isdigit()==True:
            return 0
        elif len(password)<8 and \
        password.isalpha()==True:
            return 1
        elif len(password)>7 and sum(c.isdigit() for c in password) > 0 and \
        sum(c.isalnum() for c in password) == len(password):
            return 2
        elif len(password) > 7 and \
        sum(c.isdigit() for c in password) > 0 and \
        sum(c.isalpha() for c in password) > 0 and \
        (sum(c.isalpha() for c in password) + sum(c.isdigit() for c in password)) < len(password):
            return 3
        else:
            return 4

    if passwordValidator(password)==0:
        verdict = ' very weak'
    if passwordValidator(password)==1:
        verdict = ' weak'
    if passwordValidator(password) == 2:
        verdict = ' strong'
    if passwordValidator(password)==3:
        verdict = ' very strong'
    if passwordValidator(password)==4:
        verdict = 'n invalid'

    print('The password \'{password}\' is a{verdict} password.'.format(password=password, verdict=verdict))

problem10()
```
