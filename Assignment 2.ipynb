where

- A is total alcohol consumed, in ounces (oz).
- W is body weight in pounds.
- r is the alcohol distribution ratio:
  - 0.73 for men
  - 0.66 for women
- H is number of hours since the last drink.

# Problem 11 (6 points)

Write a program that will help you determine how many months it will take to pay off a credit card balance. The program should ask the user to enter the balance of a credit card and the APR of the card. The program should then return the number of months needed.

The formula for this is

$n = -\frac{1}{30} \times \frac{\log(1 + \frac{b}{p}(1 - (1 + i)^{30}))}{\log(1+i)}$

where

- `n` is the number of months.
- `i` is the daily rate (APR divided by 365).
- `b` is the balance.
- `p` is the monthly payment.

## Example Output

<pre>
What is your balance? 5000
What is the APR on the card (as a percent)? 12
What is the monthly payment you can make? 100

It will take you 70 months to pay off this card.
</pre>

## Constraints

- Prompt for the card’s APR. Do the division internally.
- Prompt for the APR as a percentage, not a decimal.
- Use a function called `calculateMonthsUntilPaidOff`, which accepts the balance, the APR, and the monthly payment as its arguments and returns the number of months. Don’t access any of these values outside the function.
- Round fractions of a cent up to the next cent.

```python
#the problem prompts for 3 float inputs, then uses them as arguments in a months till paid off function which,
#by returning float values, allows us to calculate the logs in the formula successfully without rounding to 0
#returning this function and storing it as variable monthstopayoff allows us to call the ceiling rounded results
#in a print function to return back to the user

def problem11():
        import math
        balance = float(input('What is your balance? '))
        APR = float(input('What is the APR on the card (as a percent)? '))
        monthlypayment = float(input('What is the monthly payment you can make? '))

        def calculateMonthsUntilPaidOff(balance,APR,monthlypayment):
            import math

            dailyrate = float(APR/100/365)
            #separate formula into steps
            t1 = float(-1/30)
            t2 = float((1 + dailyrate) ** 30)
            t3= float(1-t2)
            t4= float((balance/monthlypayment) * t3)
            t5=1+t4
            t6=math.log10(t5)
            t7=math.log10(1+dailyrate)
            t8 = t6/t7
            months=t8*(-1/30) #months to pay off, final formula result

            return months

        monthstopayoff=math.ceil(calculateMonthsUntilPaidOff(balance,APR,monthlypayment)) #round off function return

        print('It will take you {monthstopayoff} months to pay off this card.'.format(monthstopayoff=monthstopayoff))

problem11()
```

```python

#this program takes for sets of strings (lowercase alpha, uppercase alpha, numeric 0 to 9 and special characters
#and stores these strings in lists. minimum length is prompted, and then the randint function adds a random
#integer between 0 and 9 to get the actual password length (greater than or equal to minimum length)
#then the program prompts the user for the number of special characters and numbers, and calculates the
#number of alphabetic characters as actual length minus special characters and number counts

#the password character list starts empty, and then the appropriate number of characters from each list
#are appended (with half uppercase and half lowercase numbers). then, the shuffle function introduces
#randomness back into the password character list to arrive at the final password

def problem22():
    import random

    alphabet = "abcdefghijklmnopqrstuvwxyz"
    upperalphabet = alphabet.upper()
    numeric = '0123456789'
    special = '!@#$%^&*'
    #storing strings in list for easier referencing based on indexes later
    alist = list(alphabet)
    ualist = list(upperalphabet)
    nlist = list(numeric)
    slist = list(special)
    minlength = int(input('What is the minimum length? '))
    #randomly adjust the minimum length by +0 to +9 to get a password at least as long as required
    actuallength = minlength + random.randint(0,9)
    specchar = int(input('How many special characters? '))
    numbers = int(input('How many numbers?'))
    #number of alphabetic characters is the actual length minus numerical and special characters
    alpha = int(actuallength - specchar - numbers)
    pwlist = []

    for i in range(specchar):
        pwlist.append(slist[random.randrange(len(slist))])
    for i in range(numbers):
        pwlist.append(nlist[random.randrange(len(nlist))])
    for i in range(alpha // 2):
        pwlist.append(alist[random.randrange(len(alist))])
    for i in range(alpha // 2):
        pwlist.append(ualist[random.randrange(len(ualist))])

    random.shuffle(pwlist)
    pwstring = "".join(pwlist)

    print(pwstring)

problem22()
```

# Problem 29 (4 points)

Given an input file, read the file and look for all occurrences of the word utilize. Replace each occurrence with use. Write the modified file to a new file.

## Example Output

Given the input file of

> One should never utilize the word "utilize" in writing. Use "use" instead.

The program should generate an output file containing

> One should never use the word "use" in writing. Use "use" instead.

## Constraints

- Prompt for the name of the output file.
- Write the output to a new file.

```python
#adapted from think python chapter 11 example
#the patternreplacement function asks for the source file name, the string pattern to match (utilize) and what string to use as a replacement (use)
#the patternreplacement function prompts for the user to enter the name of the new text file, set to be saved in the same location as the original
#the problem29 function goes through each line of the read-access original file, and replaces the utilize pattern with use with write access
#to the user input specified new file name

from __future__ import print_function, division

def problem29(pattern, replace, inputfile, outputfile):
    finputfile = open(inputfile, 'r')
    foutputfile = open(outputfile, 'w')

    for line in finputfile:
        line = line.replace(pattern, replace) #for each line in the file, look for the pattern we don't want anymore
        #and replace it with the one that we do
        foutputfile.write(line) #then, write each modified line to the new file until we've finished looping thorugh
        
    finputfile.close()
    foutputfile.close()


def patternreplacement():
    pattern = 'utilize'
    replace = 'use'
    inputfile = 'HW2P29.txt'
    outputfilename = input('Please enter the replacement file name (to be saved in the same location as the original): ')
    outputfile = '{outputfilename}.txt'.format(outputfilename=outputfilename)
    problem29(pattern, replace, inputfile, outputfile)

if __name__ == '__main__':
    patternreplacement()
```

# Problem 30 (3 points)

Create a program that reads in a file and counts the frequency of words in the file. Then construct a histogram displaying the words and the frequency, and display the histogram to the screen.

## Example Output

Given the text file words.txt with this content

> badger badger badger badger mushroom mushroom snake badger badger badger

the program would produce the following output:

<pre>
badger:   *******
mushroom: **
snake:    *
</pre>

## Constraint

- Ensure that the most used word is at the top of the report and the least used words are at the bottom.

```python
#this program reads a user-specified text file, and creates a dictionary of the unique words and their
#count of occurances within the text file. then, two lists are created from the iterated dictionary
#sortedcounts and sorted words, sorted in reverse order (highest first/on top)
#then, a while statement runs for the length of the lists and prints the unique words, a colon, and the
#number of stars corresponding to the number of times each word appears in the text file

def problem30():
    with open ('C:/Users/paigebiller/Documents/GitHub/HomeworkTwo/HW2P30.txt', 'r') as p30txt:
        file_contents = p30txt.read()
        word_list = file_contents.strip().split() #read in the entire file contents
        worddict = {word: word_list.count(word) for word in word_list}
        sortedcounts = sorted(worddict.values(), reverse=True) #sort in reverse so we display properly
        sortedwords  = sorted(worddict, key=worddict.__getitem__, reverse=True) #match sort on counts
        print(sortedcounts)
        print(sortedwords)
    histo=''
    a = 0
    while a < len(sortedcounts):
        histo += '{:>8}: {}\n'.format(sortedwords[a], '*' * (sortedcounts[a]))
        a += 1
    print(histo)
problem30()
```
